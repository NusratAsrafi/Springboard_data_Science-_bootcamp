```python
# Filter data for the last two decades (2003 to 2023)
properties_last_two_decades = properties_data[(properties_data['year'] >= 2003) & (properties_data['year'] <= 2023)]
```

```python
# Calculate the average yearly price for each borough
average_prices = properties_last_two_decades.groupby(['year', 'London_borough'])['AveragePrice'].mean().reset_index()
```

```python
# Calculate the price increase for each borough from 2003 to 2023
price_increase = average_prices.pivot(index='year', columns='London_borough', values='AveragePrice').pct_change().mean().reset_index()
price_increase.columns = ['London_borough', 'Price_Increase']
```

```python
# Sort the boroughs by the greatest price increase
sorted_price_increase = price_increase.sort_values(by='Price_Increase', ascending=False)
sorted_price_increase
```

```python
# Plot the results using Matplotlib
plt.figure(figsize=(12, 8))
plt.bar(sorted_price_increase['London_borough'], sorted_price_increase['Price_Increase'], color='skyblue')
plt.xlabel('London Borough')
plt.ylabel('Average Price Increase')
plt.title('Average Price Increase in London Boroughs (2003-2023)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

From 2003 to 2023,overall the house price increase all overs the London Borough.Moreover,the Kensington & Cheisea and Westminster boroughs has the greatest increase around 5% on average and Harrow and Croydon has 4 percent increase on averge which is lowest price increase.

```python
# Filter data for the years 1998 and 2018
properties_data_1998 = properties_data[properties_data['year'] == 1998]
properties_data_2018 = properties_data[properties_data['year'] == 2018]
```

```python
# Calculate the average price for each borough in 1998 and 2018
average_price_1998 = properties_data_1998.groupby('London_borough')['AveragePrice'].mean().reset_index()
average_price_2018 = properties_data_2018.groupby('London_borough')['AveragePrice'].mean().reset_index()
```

```python
# Merge the two dataframes on Borough
price_comparison = pd.merge(average_price_1998, average_price_2018, on='London_borough', suffixes=('_1998', '_2018'))
price_comparison
```

```python
# Calculate the ratio of 2018 prices to 1998 prices
price_comparison['Price_Ratio'] = price_comparison['AveragePrice_2018'] / price_comparison['AveragePrice_1998']
```

```python
# Sort by the ratio to see which boroughs have the highest increase
sorted_price_comparison = price_comparison.sort_values(by='Price_Ratio', ascending=False)
```

```python
# Plot the results using Matplotlib
plt.figure(figsize=(12, 8))
plt.bar(sorted_price_comparison['London_borough'], sorted_price_comparison['Price_Ratio'], color='black')
plt.xlabel('London Borough')
plt.ylabel('Price Ratio (2018 / 1998)')
plt.title('Ratio of House Prices in 2018 to 1998 for London Boroughs')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

The analysis reveals significant variations in housing price increases across London boroughs. 
The borough of Hackney shows the highest price ratio of 6.19 , indicating that average housing prices increased
more than sixfold between 1998 and 2018. Conversely, Hounshow is the lowest increase with a ratio of 3.9.
